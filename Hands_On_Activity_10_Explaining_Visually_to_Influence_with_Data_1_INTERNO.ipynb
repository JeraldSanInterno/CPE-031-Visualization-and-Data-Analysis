**4. Procedures:**

Step 1. Import Libraries and Create Sample Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample sales dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [200, 250, 180, 300, 280, 350],
    "Marketing_Spend": [50, 60, 45, 70, 65, 80]
}
df = pd.DataFrame(data)
df
```

Step 2. Know Your Audience — Compare Chart Styles

```python
# Executive audience: prefer clarity (simple bar)
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="steelblue")
plt.title("Executive View: Monthly Sales Overview", fontsize=13)
plt.show()

# Analyst audience: prefer detailed trends (line chart)
sns.lineplot(x="Month", y="Sales", data=df, marker="o", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", label="Marketing Spend")
plt.title("Analyst View: Sales vs Marketing Trend", fontsize=13)
plt.legend()
plt.show()
```

**Task 1:** Create one additional visualization suited for a different audience (e.g., a customer, investor, or teammate).

```python
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="black")
plt.title("Monthly Sales Growth from an Investor's Perspective", fontsize=13)
plt.ylabel("Sales ($)")
plt.show()
```



Step 3. Know Your Message — Highlight What Matters

```python
# Highlight the highest sales month (preattentive color cue)
highlight = df["Sales"].idxmax()

colors = ["gray" if i != highlight else "crimson" for i in range(len(df))]
plt.figure(figsize=(7,4))
plt.bar(df["Month"], df["Sales"], color=colors)
plt.title("Highlighting the Best Month in Sales", fontsize=13)
plt.show()
```

**Task 2:** Use color or annotations to emphasize a key insight (e.g., lowest point, sudden increase).

```python
lowlight = df["Sales"].idxmin()

colors = ["red" if i != lowlight else "pink" for i in range(len(df))]

plt.figure(figsize=(7,4))
plt.bar(df["Month"], df["Sales"], color=colors)
plt.title("Analysis of the Weakest Sales Period", fontsize=13)

plt.text(df["Month"][lowlight], df["Sales"][lowlight] + 50, "Lowest Point",
         ha='center', color='royalblue', fontsize=10, fontweight='bold')

plt.show()
```

```python
df["Change"] = df["Sales"].diff()

jump = df["Change"].idxmax()

colors = ["yellow" if i not in [jump, jump-1] else "orange" for i in range(len(df))]

plt.figure(figsize=(7,4))
plt.bar(df["Month"], df["Sales"], color=colors)
plt.title("Analyzing the Significant Sales Upturn", fontsize=13)

plt.annotate("Big Jump Here →",
             xy=(df["Month"][jump], df["Sales"][jump]),
             xytext=(df["Month"][jump], df["Sales"][jump] + 300))

plt.show()
```

Step 4. Storytelling with Charts — Beginning, Middle, End

```python
# Beginning: Context
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="skyblue")
plt.title("Beginning: Sales Started Steady", fontsize=13)
plt.show()

# Middle: Change or Challenge
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, color="orange", marker="o")
plt.title("Middle: A Drop in March Due to Low Marketing", fontsize=13)
plt.show()

# End: Resolution
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="limegreen")
plt.title("End: Recovery and Growth After March", fontsize=13)
plt.show()
```

Step 5. Bringing It All Together — Presentation Design

```python
# Simple combined story chart
plt.figure(figsize=(8,5))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", linewidth=2, color="royalblue", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", linestyle="--", color="orange", label="Marketing Spend")

plt.fill_between(df["Month"], df["Sales"], alpha=0.1, color="royalblue")
plt.title("From Effort to Impact: Marketing Spend Drives Sales Growth", fontsize=13)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the user-provided dataset
df = pd.read_csv('/content/drive/MyDrive/tsunami/earthquake_data_tsunami.csv')

# 1. DATA PREPARATION: Count Tsunami Outcomes
# Count the number of earthquakes that did/did not cause a tsunami (1=Yes, 0=No)
tsunami_counts = df['tsunami'].value_counts().reset_index()
tsunami_counts.columns = ['Tsunami_Caused', 'Count']

# Map the 0/1 values to descriptive labels
tsunami_counts['Tsunami_Caused'] = tsunami_counts['Tsunami_Caused'].map({
    0: 'No Tsunami',
    1: 'Tsunami Caused'
})

# 2. HIGHLIGHTING LOGIC
# Identify the highest value to highlight
max_count = tsunami_counts['Count'].max()
highlight_category = tsunami_counts.loc[tsunami_counts['Count'] == max_count, 'Tsunami_Caused'].iloc[0]

# Define colors: 'red' for the highlight, 'skyblue' for the rest
colors = ['red' if cat == highlight_category else 'skyblue'
          for cat in tsunami_counts['Tsunami_Caused']]

# 3. VISUALIZATION AND PLOT SETUP
plt.figure(figsize=(8, 6))
sns.set_style("whitegrid")
barplot = sns.barplot(x='Tsunami_Caused', y='Count', data=tsunami_counts, palette=colors)

# 4. TITLE, LABELS, AND ANNOTATION
plt.title('Count of Earthquakes That Caused a Tsunami', fontsize=16, fontweight='bold')
plt.xlabel('Tsunami Consequence', fontsize=12)
plt.ylabel('Number of Earthquakes', fontsize=12)

# Get the details for the annotation
highlight_bar_index = tsunami_counts.index[tsunami_counts['Tsunami_Caused'] == highlight_category][0]
bar_height = tsunami_counts.loc[highlight_bar_index, 'Count']

# Add the count value on top of the highlighted bar
plt.text(highlight_bar_index, bar_height + 0.1, f'{bar_height}', ha='center', va='bottom', fontsize=11, fontweight='bold')

# 5. STORYTELLING CAPTION
plt.figtext(0.5, 0.01,
            f"Story: The majority of the powerful earthquakes in this dataset ({bar_height} out of {tsunami_counts['Count'].sum()}) **resulted in a tsunami** (highlighted in red), suggesting this dataset heavily focuses on high-impact seismic events.",
            wrap=True,
            horizontalalignment='center',
            fontsize=10,
            bbox={"facecolor":"lightgrey", "alpha":0.5, "pad":5})

# Final plot adjustments
plt.ylim(0, bar_height + 2) # Adjust y-limit to fit the annotation
plt.tight_layout(rect=[0, 0.15, 1, 1])

# Display the plot
plt.show()
```

**TELL THE STORY ABOUT IT:**
A disastrous tidal wave is nearly invariably the result of intense seismic activity, as demonstrated by the focused and dramatic story of earthquakes recorded in this short dataset.  A startling 13 of the 14 big occurrences that were documented resulted in a confirmed tsunami, suggesting that this collection mostly consists of strong, high-impact seismic events with epicenters that are probably close to or beneath large bodies of water.  Given the overwhelming likelihood of a damaging follow-up wave, the data clearly demonstrates the crucial connection between high-magnitude tremors and ocean floor movement, highlighting the necessity of quick detection and warning systems in seismically active zones.  In this cohort, the one incident that did not result in a tsunami is the exception rather than the rule.
